```python
text_cols = [
    'title',
    'description',
    'input_description',
    'output_description'
]

for col in text_cols:
    df[col] = df[col].fillna("").astype(str)
```

```python
df['input_text'] = (
    df['title'] + " " +
    df['description'] + " " +
    df['input_description'] + " " +
    df['output_description']
)
```

```python
df['input_text'].iloc[5][:500]
```

```python
import re

def clean_text(text):
    text = text.lower()
    text = re.sub(r'<.*?>', ' ', text)     # remove HTML
    text = re.sub(r'\s+', ' ', text)       # normalize spaces
    return text.strip()
```

```python
df['clean_text'] = df['input_text'].apply(clean_text)
```

```python
print(df['input_text'][0][:300])
print(df['clean_text'][0][:300])
```

```python
df['text_length'] = df['clean_text'].apply(len)
df['word_count'] = df['clean_text'].apply(lambda x: len(x.split()))
```

```python
keywords = [
    'graph', 'tree', 'dp', 'dynamic programming',
    'recursion', 'bfs', 'dfs',
    'greedy', 'binary search',
    'segment tree', 'modulo'
]
for kw in keywords:
    df[f'kw_{kw.replace(" ", "_")}'] = df['clean_text'].apply(
        lambda x: x.count(kw)
    )
```

```python
df.groupby('problem_class')['kw_dp'].mean()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['class_label'] = le.fit_transform(df['problem_class'])
```

```python
dict(zip(le.classes_, le.transform(le.classes_)))
```

```python
cols_req = (
    ['clean_text', 'text_length', 'word_count'] +
    [col for col in df.columns if col.startswith('kw_')] +
    ['class_label', 'problem_score']
)
df_final=df[cols_req]
```
